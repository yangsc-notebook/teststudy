```python
#多次显示图表
from pyecharts import Bar,Line
from pyecharts.engine import create_default_environment
bar=Bar("这是第一个标题","这是第二个标题")
bar.add("服装",["衬衫","羊毛衫","雪纺衫","裤子","高跟鞋","袜子"],[5,20,36,10,75,90])

line=Line("这是第一个标题","这是第二个标题")
line.add("服装",["衬衫","羊毛衫","雪纺衫","裤子","高跟鞋","袜子"],[5,20,36,10,75,90])

env=create_default_environment("html")
#为渲染创建一个配置环境
#create_default_environment(filetype)
#file_type:'html','svg','png','jpeg','gif','pdf'
env.render_chart_to_file(bar,path="bar.html")
env.render_chart_to_file(line,path="line.html")
```

```python
import pandas as pd
import numpy as np
from pyecharts import Bar
```

```python
title=Bar("bar chart","1111")
index=pd.date_range('3/8/2017',periods=6,freq='M')  #增加月份
df1=pd.DataFrame(np.random.randn(6),index=index)
df2=pd.DataFrame(np.random.randn(6),index=index)
dtvalue1=[i[0] for i in df1.values]
dtvalue2=[i[0] for i in df2.values]
_index=[i for i in df1.index.format()]
bar.add('profit',_index,dtvalue1)
bar.add('loss',_index,dtvalue2)
bar.render("bar_chart.html")
```

```python
from pyecharts import Bar
bar=Bar("柱状图")
attr=["{}月".format(i) for i in range(1,13)]
v1=[2.0,4.9,7.0,23.2,25.6,76.7,135.6,162.2,32.6,20.0,6.4,3.3]
v2=[2.6,5.9,9.0,26.4,28.7,70.7,175.6,182.2,48.7,18.8,6.0,2.3]
bar.add("蒸发量",attr,v1,mark_line=["average"],mark_point=["max","min"])
bar.add("降雨量",attr,v2,mark_line=["average"],mark_point=["max","min"])
bar.render("xiyu.html")
```

```python
#饼状图
from pyecharts import Pie
pie=Pie("饼图-玫瑰图实例",title_pos='center',width=900) #标题
pie.use_theme("dark")
attr=["衬衫","羊毛衫","雪纺衫","裤子","高跟鞋","袜子"]
v1=[11,12,13,10,10,10]
v2=[19,21,32,20,20,33]
#is_legend_show=True:鼠标接触显示
#is_label_show=True 直接显示
pie.add("商品A",attr,v1,center=[25,50],is_random=True,radius=[30,75],rosetype='radius')
pie.add("商品B",attr,v2,center=[75,50],is_random=True,radius=[30,75],rosetype='area',
       is_legend_show=False,is_label_show=True)
pie
```

```python
from pyecharts import Bar
bar=Bar("柱状图堆叠显示")
bar.add("a",attr,v1,is_stack=True)
bar.add("b",attr,v2,is_stack=True)
bar
```

```python
#使用标记点和标记线
from pyecharts import Bar
bar=Bar("标记线和标记点实例")
bar.add("商家A",attr,v1,mark_point=["average"])
bar.add("商家B",attr,v2,mark_line=["min","max"])
bar
```

```python
#is_convert交换XY轴
from pyecharts import Bar
attr=["衬衫","羊毛衫","雪纺衫","裤子","高跟鞋","袜子"]
v1=[11,12,13,10,10,10]
v2=[19,21,32,20,20,33]
bar=Bar("x轴和y轴交换")
bar.add("商家A",["衬衫","羊毛衫","雪纺衫","裤子","高跟鞋","袜子"],v1)
bar.add("商家B",["衬衫","羊毛衫","雪纺衫","裤子","高跟鞋","袜子"],v2, is_convert=True)
bar
```

```python
#dataZoom效果，slider类型//区域缩放
import random
attr=["{}天".format(i+1) for i in range(30)]
v1 = [random.randint(1,30) for _ in range(30)]
bar = Bar("bar-datazoom-slider")
bar.add("",attr,v1,is_label_show=True,is_datazoom_show=True)
bar.render("缩放.HTML")
```

```python
#datazoom效果 inside类型
attr=["{}天".format(i) for i in range(30)]
v1=[random.randint(1,30) for _ in range(30)]
bar=Bar("bar-datazoom-inside")
bar.add("",attr,v1,is_datazoom_show=True,datazoom_type="inside",datazoom_range=[10,25])
bar.render("datazoom_inside.html")
```

```python
#多datazoom效果，效果同时支持X,Y轴
days=["{}天".format(i) for i in range(30)]
days_v1=[random.randint(1,30) for _ in range(30)]
bar=Bar("bar-datazoom-xy")
bar.add("",days,days_v1,
       #默认是X轴横向
        is_datazoom_show=True,
        datazoom_type="slider",
        datazoom_range=[10,25],
        #新增额外的datazoom控制条，纵向
        is_datazoom_extra_show=True,
        datazoom_extra_type="slider",
        datazoom_extra_range=[10,25],
        is_toolbox_show=False
       )
bar.render("xy双向控制.html")
```
