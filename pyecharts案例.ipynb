```python
import numpy as np
import pandas as pd
import pyecharts
```

```python
data1=pd.read_csv("/home/ysc/数据分析/data/第九讲/qiushi_info.csv",encoding='utf-8')
data1.head()
```

```python
data2=pd.read_csv("/home/ysc/数据分析/data/第九讲/user_info.csv",encoding='utf-8')
data2.head()
```

```python
#对段子数据进行清洗
data1.info()
```

```python
data1['age'].unique()
```

```python
#年龄处理方法：将不详替换成0，age转为int，平均值替换
data1['age'].replace("不详",0,inplace=True)
data1['age']=data1['age'].astype('int64')

data1['age'].replace(0,int(data1[data1['age']!=0]['age'].mean()),inplace=True)
```

```python
data1.dtypes
```

```python
#对用户数据进行处理，查看缺失值，5个缺失值删除
data2.isnull().sum()
```

```python
#使用dropna删除缺失值，将浮点型转为整型
data2.dropna(inplace=True)

data2['fans']=data2['fans'].astype('int64')
data2['topic']=data2['topic'].astype('int64')
data2['qiushi']=data2['qiushi'].astype('int64')
data2['comment_1']=data2['comment_1'].astype('int64')
data2['favour']=data2['favour'].astype('int64')
data2['handpick']=data2['handpick'].astype('int64')
```

```python
#对于home为了数据可视化，通过字符串增加一列，用于获取省份的数据
data2['province']=data2['home'].str.split('·').str[0]
data2.head()
```

```python
#用户的糗事年龄删除天转为int型
data2['qiushi_age']=data2['qiushi_age'].str.strip('天')
data2['qiushi_age']=data2['qiushi_age'].astype('int64')
```

```python
#数据探索--分析好笑数和评论多的段子是那些用户发的
laugh_sort=data1.sort_values(by='laugh',ascending=False)[0:10]
laugh_sort
```

```python
bar=pyecharts.Bar("搞笑段子用户排名")
attr=list(laugh_sort['id'])
v1=list(laugh_sort['laugh'])
bar.add("好笑数",attr,v1,is_label_show=True,mark_point=['max','min'],
        mark_line=['average'],is_xaxislabel_align=True,xaxis_interval=0,
        xaxis_name_size=12,xaxis_rotate=30)
#is_xaxislabel_align=True,为x轴刻度与x轴标签对应
#xaxis_interval=0, 设置每个x标签都显示
#xaxis_name_size=12,设置x标签大小
#xaxis_rotate=30设置x轴标签旋转
bar
```

```python
#对comment字段进行排序，画图
comment_sort=data1.sort_values(by='comment',ascending=False)[0:10]
comment_sort
```

```python
bar=pyecharts.Bar("评论段子用户排名")
attr=list(comment_sort["id"])
v1=list(comment_sort["comment"])
bar.add("评论数",attr,v1,is_label_show=True,mark_point=['max','min'],
        mark_line=['average'],is_xaxislabel_align=True,xaxis_interval=0,
       xaxis_name_size=12,xaxis_rotate=30)

#is_xaxislabel_align=True,为x轴刻度与x轴标签对应
#xaxis_interval=0, 设置每个x标签都显示
#xaxis_name_size=12,设置x标签大小
#xaxis_rotate=30设置x轴标签旋转
bar
```

```python
#删除data1中缺失值，合并data1和data2，删除重复值
data1=data1.dropna() #删除缺失值
```
