# 数据库的读取与存储

```python
#连接数据库
import pymysql
conn=pymysql.connect(
                     host='localhost',user='root',passwd='123456',
                     db='mystudy',port=3306,charset='utf8')
```

```python
#创建游标
cursor=conn.cursor()
#创建table表
creat='''create table ch3ex9(
       id int,
       name char(8),
       grade int
       )engine innodb default charset=utf8;'''
cursor.execute(creat)  #执行命令
conn.commit()          #完成命令
```

```python
#mysql添加数据
cursor.execute("insert into ch3ex9 (id,name,grade) values(%s,%s,%s)",(1,'luchy',87))
cursor.execute("insert into ch3ex9 (id,name,grade) values(%s,%s,%s)",(2,'peter',92))
cursor.execute("insert into ch3ex9 (id,name,grade) values(%s,%s,%s)",(3,'lili',85))
conn.commit()
cursor.close()  #关闭游标
```

```python
#读取数据库内容
import pandas as pd
df=pd.read_sql('select* from ch3ex9',conn)
df
```
