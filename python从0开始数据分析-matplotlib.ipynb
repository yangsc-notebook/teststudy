# 柱状图

```python
data=[23,85,72,43,52]
plt.bar([1,2,3,4,5],data)
```

```python
plt.bar(range(len(data)),data,color='royalblue',alpha=0.7)
plt.grid(color='#95a5a6',linestyle='--',linewidth=1,axis='y',alpha=0.5)
```

```python
data1=[23,85,72,43,52]
data2=[42,35,21,16,9]
plt.bar(range(len(data)),data1)
plt.bar(range(len(data)),data2,bottom=data1)
```

```python
#绘制并列柱状图
data1=[23,85,72,43,52]
data2=[42,35,21,16,9]
width=0.3
plt.bar(np.arange(len(data1)),data1,width=width,label='one')
plt.bar(np.arange(len(data2))+width,data2,width=width,label='two')
plt.legend()
```

```python
#
data3=[23,85,72,43,52]
plt.barh(range(len(data3)),data3)
```

```python
#刻度与标签
data=[23,85,72,43,52]
labels=['A','B','C','D','E']
plt.xticks(range(len(data)),labels)  #设置刻度和标签
plt.bar(range(len(data)),data)
```

```python
data=[23,72,85,43,52]
labels=['A','B','C','D','E']
plt.xticks(range(len(data)),labels)#设置刻度和标签
plt.xlabel('Class')
plt.ylabel('Amounts')
plt.title('Example')
plt.bar(range(len(data)),data)
```

```python
country_count=starbucks['Country'].value_counts()[0:10]
country_count
```

```python
plt.style.use('ggplot') #设置图表样式
labels=list(country_count.index) #刻度标签
plt.xlabel('country')
plt.ylabel('count')
plt.title('country top 10')
plt.bar(range(len(labels)),country_count)
plt.xticks(range(len(labels)),labels,fontsize=12)
```

```python
city_count =starbucks['City'].value_counts()[0:10]
city_count
```

```python
plt.figure(figsize=(10,6))
plt.rcParams['font.sans-serif']=[u'Simhei']
plt.rcParams['axes.unicode_minus']=False

labels=list(city_count.index)
plt.xlabel('city')
plt.ylabel('count')
plt.title('city top 10')
plt.bar(range(len(labels)),city_count)
plt.xticks(range(len(labels)),labels)
```
